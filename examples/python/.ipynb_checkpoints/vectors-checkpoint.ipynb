VECTOR MULTIPLICATION

```python
v1 = np.array([2, 5, 7])
v2 = np.array([4, 6, 1])

v3 = np.dot(v1,v2)

print(v1, v2, v3)
```

```python
v3 = np.matmul(v1, v2)
print(v3)
```

## Hadamard Product

```python
print(v1*v2) # element wise multiplication 
print(np.multiply(v1,v2))
```

## Cross Product

![caption](cross_product.png)

```python
v3 = np.cross(v1,v2) # cross product
print(v3)
```

```python
fig = plt.figure()
ax = fig.gca(projection = '3d')
xx, yy = np.meshgrid(np.linspace(-20,20,20), np.linspace(-20,20,20))
z1 = (-v3[0]*xx - v3[1]*yy)/v3[2]
ax.plot_surface(xx,yy,z1)

## pot the two vectors
ax.plot([0,v1[0]], [0,v1[1]], [0,v1[2]], 'k')
ax.plot([0,v2[0]], [0,v2[1]], [0,v2[2]], 'k')
ax.plot([0,v3[0]], [0,v3[1]], [0,v3[2]], 'r')

ax.view_init(azim=150, elev = 45)
plt.show()
```

## Hermitian Transpose

```python
v3 = np.array([np.complex(3,4), np.complex(2,3)])  # 3 + 4i
print(np.transpose(np.conjugate(v3))) # Hermitian : transpose conjugate 
print(np.transpose(v3))

print(v3[:, np.newaxis]) # print it as a column vector
print(v3[np.newaxis, :]) # print it as a row vector again 

print( np.linalg.norm(v3[0]))  # (3 -4i)(3 + 4i)
```

## Unit Vector

```python
v3 = [ 2, 5]
mu = 1/np.linalg.norm(v3)
print(mu)
```

```python
unitvector = np.dot(v3,mu)
print('Univector : v3/||v3|| is', unitvector)
print('The norm of a unit vector is: ', np.linalg.norm(unitvector)) # norm of a unit vector is one 
```
